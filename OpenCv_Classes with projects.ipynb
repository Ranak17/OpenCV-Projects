# Project 2 : Document scanner

```python
import cv2
import numpy as np
def preprocessing(img):
    gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    blur_img=cv2.GaussianBlur(gray_img,(7,7),1)
    canny_img=cv2.Canny(blur_img,100,100)
    kernel=np.ones((5,5))
    dial_img=cv2.dilate(canny_img,kernel,iterations=2)
    threshold_img=cv2.erode(dial_img,kernel,iterations=1)
#     cv2.imshow("threshold_img",threshold_img)
    return threshold_img

def get_contours(img,contour_img):
    try:
        contours,hierarchy=cv2.findContours(img,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
        biggest=0
        Maxarea=0
        for cnt in contours:
            area=cv2.contourArea(cnt)
            if area>=10000:
    #             cv2.drawContours(contour_img,cnt,-1,(0,255,0),2)
                peri=cv2.arcLength(cnt,True)
                approx=cv2.approxPolyDP(cnt,0.02*peri,True)
                if area>Maxarea and len(approx)==4:
                    biggest=approx
                    print(biggest)
                    Maxarea=area
        cv2.drawContours(contour_img,biggest,-1,(0,255,0),10)
    #     print(biggest.shape)
        return biggest
    except:
        return np.array([])


def reorder(mypoints):
    mypoints=mypoints.reshape((4,2))
    mynewpoint=np.zeros((4,1,2),np.int32)
    add=mypoints.sum(1)
    mynewpoint[0]=mypoints[np.argmin(add)]
    mynewpoint[3]=mypoints[np.argmax(add)]
    diff=np.diff(mypoints,axis=1)
#     print(diff)
    mynewpoint[1]=mypoints[np.argmin(diff)]
    mynewpoint[2]=mypoints[np.argmax(diff)]
    
#     print("add",add)
    return mynewpoint


def getwarp(img,biggest):
    biggest=reorder(biggest)
    pts1=np.float32(biggest)
    pts2=np.float32([[0,0],[width_img,0],[0,height_img],[width_img,height_img]])
    matrix=cv2.getPerspectiveTransform(pts1,pts2)
    warp_img=cv2.warpPerspective(img,matrix,(width_img,height_img))
    return warp_img

width_img=480
height_img=640



cap=cv2.VideoCapture("http://26.159.20.114:8080/video")
# cap.set(3,1640)#width
# cap.set(4,16400)#height
# cap.set(10,1)#brightne
while True:
    try:
        success,img=cap.read()
        contour_img=img.copy()
    #     img=cv2.resize(img,(width_img,height_img))
    #     cv2.imshow("ds",img)
        pre_processed_img=preprocessing(img)
        biggest=get_contours(pre_processed_img,contour_img)
        print(biggest.size)
        if biggest.size!=0:
            warped_img=getwarp(img,biggest)
            cropped_img=warped_img[20:warped_img.shape[0]-20,20:warped_img.shape[1]-20]
            cv2.imshow("warped_image",cropped_img)
            cv2.imshow("result",pre_processed_img)
            cv2.imshow("contour_img",contour_img)
        else:
            cv2.imshow("contour_img",contour_img)
            cv2.imshow("result",pre_processed_img)
        if cv2.waitKey(1) & 0xFF ==ord('q'):
            break
    except:
        pass
cap.release()
cv2.destroyAllWindows()
```

# Project 3: Number plate detection 

```python
import cv2
cap=cv2.VideoCapture("http://26.159.20.114:8080/video")
# cap.set(3,1640)#width
# cap.set(4,16400)#height
# cap.set(10,1)#brightne


nplatecascade=cv2.CascadeClassifier(r"C:\Users\karan\anaconda3\Lib\site-packages\cv2\data\haarcascade_russian_plate_number.xml")
while True:
    success,img=cap.read()
    gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    nplate=nplatecascade.detectMultiScale(gray_img,1.1,4)
    cv2.imshow("image",img)
    for (x,y,w,h) in nplate:
        area=w*h
        if area>500:  
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
            cv2.putText(img,"Number plate",(x,y-5),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)         
            cv2.imshow("image",img)
    
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
